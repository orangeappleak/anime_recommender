## **Extracting weights from model**

```python
def extract_weights(name, model):
    weight_layer = model.get_layer(name)
    weights = weight_layer.get_weights()[0]
    weights = weights / np.linalg.norm(weights, axis = 1).reshape((-1, 1))
    return weights

anime_weights = extract_weights('anime_embedding', model)
user_weights = extract_weights('user_embedding', model)

id_weights = {}
for i, id in enumerate(anime2anime_encoded):
    id_weights[id] = anime_weights[i]
    
import pickle

# Save id_weights mapping
with open('id_weights.pkl', 'wb') as file:
    pickle.dump(id_weights, file)

```

### **anime meta data**

```python
df = pd.read_csv('./anime_data/anime.csv', low_memory=True)
df = df.replace("Unknown", np.nan)
```

```python
# Fixing Names
def getAnimeName(anime_id):
    try:
        name = df[df.anime_id == anime_id].eng_version.values[0]
        if name is np.nan:
            name = df[df.anime_id == anime_id].Name.values[0]
    except:
        print('error')
    
    return name

df['anime_id'] = df['MAL_ID']
df["eng_version"] = df['English name']
df['eng_version'] = df.anime_id.apply(lambda x: getAnimeName(x))

df.sort_values(by=['Score'], 
               inplace=True,
               ascending=False, 
               kind='quicksort',
               na_position='last')

df = df[["anime_id", "eng_version", 
         "Score", "Genres", "Episodes", 
         "Type", "Premiered", "Members"]]
```

```python
def getAnimeFrame(anime):
    if isinstance(anime, int):
        return df[df.anime_id == anime]
    if isinstance(anime, str):
        return df[df.eng_version == anime]
    
getAnimeFrame("One Piece")
```

```python
# # #pd.reset_option('all')
pd.set_option("max_colwidth", None)

def find_similar_animes(mal_id, n=10, return_dist=False, neg=False):
    try:
        dists = np.dot(anime_weights, id_weights[mal_id])
        sorted_dists = np.argsort(dists)
        
        n = n + 1            
        
        if neg:
            closest = sorted_dists[:n]
        else:
            closest = sorted_dists[-n:]

        print('animes closest to {}'.format(mal_id))

        if return_dist:
            return dists, closest
        
        rindex = df

        SimilarityArr = []
        keys = list(id_weights.keys())

        for close in closest:
            
            decoded_id = keys[close]
            sypnopsis = getSypnopsis(decoded_id)
            anime_frame = getAnimeFrame(decoded_id)
            
            
            anime_name = anime_frame.eng_version.values[0]
            genre = anime_frame.Genres.values[0]
            similarity = dists[close]
            SimilarityArr.append({"anime_id": decoded_id, "name": anime_name,
                                  "similarity": similarity,"genre": genre,
                                  'sypnopsis': sypnopsis})

        Frame = pd.DataFrame(SimilarityArr).sort_values(by="similarity", ascending=False)
        return Frame[Frame.anime_id != mal_id].drop(['anime_id'], axis=1)

    except:
        print('{}!, Not Found in Anime list'.format(mal_id))
```
